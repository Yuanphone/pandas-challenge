```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete.head()
```

```python
# Calculate the total number of schools
totalschools=len(school_data_complete["school_name"].unique())
# Count the total number of students
totalstudents=school_data_complete["Student ID"].count()
# Calculate the total budget
totalbudget=school_data["budget"].sum()
# Calculate the avarage math score
averagemath=school_data_complete["math_score"].mean()
# Calculate the avarage reading score
averageread=school_data_complete["reading_score"].mean()
# Calculate the % passing math
passmath=len(school_data_complete[school_data_complete["math_score"]>=70])*100/totalstudents
# Calculate the % passing reading
passread=len(school_data_complete[school_data_complete["reading_score"]>=70])*100/totalstudents
# Calculate the counts of students who passed math and reading
passboth=len(school_data_complete[(school_data_complete["math_score"]>=70)&\
                            (school_data_complete["reading_score"]>=70)])*100/totalstudents
# Create distirct summary dataframe
district_dict={"Total Schools":[totalschools],"Total Students":[totalstudents],"Total Budget":[totalbudget],\
               "Average Math Score":[averagemath],"Average Reading Score":[averageread],"% Passing Math":[passmath],\
               "% Passing Reading":[passread],"% Overall Passing":[passboth]}
District_df=pd.DataFrame(district_dict)
District_df["Total Budget"]=District_df["Total Budget"].map("${:,.2f}".format,na_action='ignore')
District_df["Total Students"]=District_df["Total Students"].map("{:,}".format)


District_df
```

```python
# Group by "school_name","type",and "budget"
grouped_school_df=school_data_complete.groupby([ "school_name","type","budget",])
# Calculate the total stuents for each school
totalstudents=grouped_school_df["Student ID"].count()
# Calculate the average math score for each school
averagemath=grouped_school_df["math_score"].mean()
# Calculate the average reading score for each school
averageread= grouped_school_df["reading_score"].mean()
# Calculate the % passing math for each school
passmath=school_data_complete.loc[school_data_complete["math_score"] >= 70].groupby(["school_name"])["math_score"].count()
percentpassmath=passmath/totalstudents*100
# Calculate the % passing reading for each school
passread=school_data_complete.loc[school_data_complete["reading_score"] >= 70].groupby(["school_name"])["reading_score"].count()
percentpassread=passread/totalstudents*100
# Calculate the counts of students who passed math and reading for each shchool
passboth=school_data_complete[(school_data_complete["math_score"] >= 70)& (school_data_complete["reading_score"]>=70)].groupby(["school_name"])["math_score"].count()
passboth
```

```python
# Create the incomplete dateframe for school summary 
school_dict={"Total Students":totalstudents,\
               "Average Math Score":averagemath,"Average Reading Score":averageread,"% Passing Math":percentpassmath,\
               "% Passing Reading":percentpassread}
# Use reset_index() to convert groupby indexes ("school_name","type" and "budget") to columns
draft_school_df=pd.DataFrame(school_dict).reset_index()
draft_school_df
```

```python
# Use merge to add the new column "passboth", but the column is under the name "math_score" which needs to be renamed
school_data_merge = pd.merge(draft_school_df, passboth, how="left", on=["school_name", "school_name"])
# Set school_name as the index
school_data_merge=school_data_merge.set_index("school_name")
school_data_merge
```

```python
# Rename columns
school_data_renamed1=school_data_merge.rename(columns={"type":"School Type",\
                                                    "budget":"Total School Budget","math_score":"Passbothcounts"},\
                                           inplace=False)
school_data_renamed1
```

```python
# Calculate the per student budget and add it as a new column
school_data_renamed1.loc[:,"Per Student Budget"]=school_data_renamed1.loc[:,"Total School Budget"]\
                                            /school_data_renamed1.loc[:,"Total Students"]
# Calculate the % overall passing and add it as a new column
school_data_renamed1.loc[:,"% Overall Passing"]=school_data_renamed1.loc[:,"Passbothcounts"]\
                                          /school_data_renamed1.loc[:,"Total Students"]*100
# Create the dataframe of the final school summary
school_data_final=school_data_renamed1.loc[:,["School Type","Total Students","Total School Budget",\
                                              "Per Student Budget", "Average Math Score",\
               "Average Reading Score","% Passing Math", "% Passing Reading","% Overall Passing"]]
# Add format
school_data_final.loc[:,"Total School Budget"]=school_data_final.loc\
                                        [:,"Total School Budget"].map("${:,.2f}".format,na_action='ignore')
school_data_final.loc[:,"Per Student Budget"]=school_data_final.loc\
                                        [:,"Per Student Budget"].map("${:,.2f}".format,na_action='ignore')
# Remove the index name "school_name"
school_data_final.index.names=['']
# Final school summary 
school_data_final
```

```python
# Highest-performing schools (by % overall passing)
school_data_final.sort_values("% Overall Passing", ascending=False).head()
```

```python
# Lowest-performing schools (by % overall passing)
school_data_final.sort_values("% Overall Passing").head()
```

```python
# math scores by Grade
mathbygrade=pd.pivot_table(school_data_complete, values="math_score", index="school_name", columns="grade",sort=True) 
mathbygrade_final=mathbygrade.loc[:,["9th","10th","11th","12th"]]
# remove the names of index and columns
mathbygrade_final.index.names=['']
mathbygrade_final.columns.names=['']
mathbygrade_final
```

```python
# Reading scores by grade
readbygrade=pd.pivot_table(school_data_complete, values="reading_score", index="school_name", columns="grade",sort=True)
readbygrade_final=mathbygrade.loc[:,["9th","10th","11th","12th"]]
# remove the names of index and columns
readbygrade_final.index.names=['']
readbygrade_final.columns.names=['']
readbygrade_final
```

```python
# Scores by school spending
bins1=[0,585,630,645,680]
spending_labels=["<$585","$585-630","$630-645","$645-680"]                                                    
school_data_final["Per Student Budget"] = school_data_final["Per Student Budget"].str.replace("$", "", regex=True)
school_data_final["Per Student Budget"] =school_data_final["Per Student Budget"].astype(float)  
school_data_final["Spending Ranges (Per Students)"] = pd.cut(school_data_final["Per Student Budget"],\
                                                             bins1,labels=spending_labels)
school_spending_grouped=school_data_final.groupby("Spending Ranges (Per Students)").mean()
school_spending_final=round(school_spending_grouped.loc[:,["Average Math Score",\
                            "Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]],2)

school_spending_final
```

```python
# Scores by school size
bins2=[0,1000,2000,5000]
size_labels=["Small(<1000)","Medium(1000-2000)","Large(2000-5000)"]                                                      
school_data_final["School Size"] = pd.cut(school_data_final["Total Students"], bins2,labels=size_labels)
school_size_grouped=school_data_final.groupby("School Size").mean()
school_size_final=school_size_grouped.loc[:,["Average Math Score",\
                        "Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]]
school_size_final
```

```python
# Scores by school types
school_type_grouped=school_data_final.groupby("School Type").mean()
school_type_final=school_type_grouped.loc[:,["Average Math Score",\
                        "Average Reading Score","% Passing Math","% Passing Reading","% Overall Passing"]]
school_type_final
```
